```python
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
def clean(txt):
    txt=txt.lower()            
    txt=re.sub(r'[^a-zA-Z]', ' ', txt)  
    txt=re.sub(r'\s+', ' ', txt)
    ntxt=word_tokenize(txt)
    stop_words=set(stopwords.words('english'))
    stemm=PorterStemmer()
    ntxt=[stemm.stem(word) for word in ntxt if word not in stop_words]
    return " ".join(ntxt)
```

```python
text_data = [

"NLP is amazing! Learning models and patterns.",

"Learning NLP is fun and powerful.",

"Deep learning improves NLP applications.",

"Text data requires cleaning and preprocessing.",

"Natural Language Processing is widely used in AI.",

"AI models learn from large amounts of text data.",

"Preprocessing text is an important step in NLP.",

"Machine learning and deep learning are related fields."

]
cleanedtxt=list(map(clean,text_data))
for new in cleanedtxt:
 print(new)
newl=[]
newl.append(new)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
count=cv.fit_transform(newl)
print(count.toarray())
print(cv.vocabulary_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
tfv=tf.fit_transform(newl).toarray()
tfv
```
