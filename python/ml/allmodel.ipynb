```python
df=pd.read_csv("housing (2).csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
bd=df["total_bedrooms"].mean()
fbd=df[df["total_bedrooms"].isna()].index
df.loc[fbd,"total_bedrooms"]=bd
```

```python
df.isna().sum()
```

```python
label=LabelEncoder()
```

```python

op=label.fit_transform(df.loc[:,["ocean_proximity"]])
df.loc[:,"ocean_proximity"]=op
```

```python
df
```

```python
sc=StandardScaler()
```

```python

x=df.drop(columns="median_house_value",axis=1)
y=df["median_house_value"]
sx=sc.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(sx,y,random_state=39,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
r2_score(ytest,pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(max_depth=10,min_samples_split=60,min_samples_leaf=10,max_features=10,random_state=42)
dt.fit(xtrain,ytrain)
```

```python
predi=dt.predict(xtest)
```

```python
r2_score(ytest,predi)*100
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
parameters = {
    "n_estimators": [50, 100, 150],
    "max_depth": [3, 5, 7],
    "criterion": ["squared_error", "absolute_error", "friedman_mse"],
    "min_samples_split": [2, 3, 4]
}

grid_search=GridSearchCV(estimator=rand,param_grid=parameters,cv=5,scoring="r2",verbose=1)
```

```python
grid_search.fit(xtrain,ytrain)
grid_search.best_params_
grid_search.best_score_
```

```python
parameters_random={
    'n_estimators':randint(40,150),
    'max_depth':randint(1,15),
    'criterion':["squared_error","absolute_error"],

}

random_search=RandomizedSearchCV(estimator=rand,cv=5,n_iter=15,scoring="r2",verbose=1,param_distributions=parameters_random)

random_search.fit(xtrain,ytrain)
```
