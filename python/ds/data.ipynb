```python
import pandas as pd

# Base district list (official dataset from DataMeet)
url = "https://raw.githubusercontent.com/datameet/india-districts/master/data/districts.csv"
df = pd.read_csv(url)

# Normalize column names
df = df.rename(columns={"state": "State", "district": "District"})

# ---- Earthquake Risk ----
high_eq_states = [
    "Himachal Pradesh","Uttarakhand","Jammu and Kashmir","Ladakh","Assam",
    "Arunachal Pradesh","Manipur","Mizoram","Nagaland","Tripura","Meghalaya",
    "Bihar","Gujarat","Delhi","Sikkim"
]

def earthquake_risk(state):
    if state in high_eq_states:
        return "Very High"
    elif state in ["Maharashtra","Punjab","Haryana","West Bengal","Rajasthan"]:
        return "High"
    elif state in ["Tamil Nadu","Karnataka","Andhra Pradesh","Telangana","Odisha"]:
        return "Moderate"
    else:
        return "Low"

# ---- Flood Risk ----
flood_high_states = [
    "Assam","Bihar","Uttar Pradesh","West Bengal","Odisha","Kerala","Jharkhand"
]

def flood_risk(state):
    if state in flood_high_states:
        return "Very High"
    elif state in ["Punjab","Haryana","Gujarat","Maharashtra","Andhra Pradesh","Tamil Nadu"]:
        return "High"
    elif state in ["Karnataka","Chhattisgarh","Madhya Pradesh"]:
        return "Moderate"
    else:
        return "Low"

# ---- Cyclone Risk ----
cyclone_very_high = ["Odisha","Andhra Pradesh","West Bengal","Tamil Nadu"]
cyclone_high = ["Gujarat","Maharashtra","Kerala","Karnataka","Goa"]

def cyclone_risk(state):
    if state in cyclone_very_high:
        return "Very High"
    elif state in cyclone_high:
        return "High"
    elif state in ["Telangana","Chhattisgarh"]:
        return "Moderate"
    else:
        return "Low"

# Apply rules
df["Earthquake_Risk"] = df["State"].apply(earthquake_risk)
df["Flood_Risk"] = df["State"].apply(flood_risk)
df["Cyclone_Risk"] = df["State"].apply(cyclone_risk)

# Save to CSV
df[["State","District","Earthquake_Risk","Flood_Risk","Cyclone_Risk"]].to_csv(
    "India_All_Districts_Disaster_Risk.csv", index=False
)

print("✅ CSV generated: India_All_Districts_Disaster_Risk.csv")
```
