```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cleaned_data.csv")
```

```python
osx=df[df["os"].isna()].index
```

```python
os=df["os"].mode()
df.loc[osx,"os"]= "Android v12"

```

```python
battery=df["battery_mah"].mode()
bat=df[df["battery_mah"].isna()].index
df.loc[bat,"battery_mah"]=battery
```

```python
df.drop(columns="clock_ghz",axis=1,inplace=True)
df.drop(columns="memory_card_max_gb",axis=1,inplace=True)
df.drop(columns="memory_card_supported",axis=1,inplace=True)
df.drop(columns="memory_card_type",axis=1,inplace=True)
df.drop(columns="VoLTE",axis=1,inplace=True)
df.drop(columns="refresh_rate_hz",axis=1,inplace=True)
df.drop(columns="sim_type",axis=1,inplace=True)
df.drop(columns="display_type",axis=1,inplace=True)

```
